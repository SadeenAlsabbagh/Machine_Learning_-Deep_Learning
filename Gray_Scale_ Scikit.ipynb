```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans 
from skimage import io 
from sklearn.preprocessing import StandardScaler


img= io.imread("/Users/alsabbaghs/Downloads/Lena-original-gray (1).png", as_gray= True)

X = img.reshape(-1,1)

kmeans = KMeans(n_clusters=2)

kmeans.fit(X)

labels = kmeans.labels_
print(min(labels))
print(max(labels))
centroids = kmeans.cluster_centers_
print(centroids)
print(X.shape)

labels = labels.reshape(img.shape)



plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.subplot(122)
plt.imshow(labels, cmap="gray")
plt.show()

```

```python
import numpy as np

np.random.seed(1000)

n = 38
X_morning = np.random.uniform(low=.02, high=.18, size=38)
X_afternoon = np.random.uniform(low=.05, high=.20, size=38)
X_night = np.random.uniform(low=.025, high=.175, size=38)
X = np.vstack([X_morning, X_afternoon, X_night]).T


from sklearn.cluster import KMeans
k = 2
kmeans = KMeans(n_clusters=k, random_state=0).fit(X)

idx = np.argsort(kmeans.cluster_centers_.sum(axis=1))
lut = np.zeros_like(idx)
lut[idx] = np.arange(k)
```

```python
from skimage import data, io
import numpy as np

painting=plt.imread("/Users/alsabbaghs/Downloads/Lena-original-gray (1).png")
gray_image = color.rgb2gray(painting)
gray_image = np.array(gray_image) * 255
gray_image = gray_image.astype('uint8')

thresh = 120
binary_image = gray_image >= thresh

print("Threshold value", thresh)
io.imshow(binary_image)
io.show()

print("gray value", gray_image)
print("\n")
print("binary value", binary_image)
```
