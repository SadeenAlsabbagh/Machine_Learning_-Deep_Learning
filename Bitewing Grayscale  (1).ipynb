```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import cv2
from sklearn.cluster import KMeans 
from skimage import io 
from sklearn.preprocessing import StandardScaler
from split_image import split_image

# original image
img = io.imread("/Users/votri/Downloads/DSDA 385 Machine Learning/KMeans Bitewing/BITEWING-XRAYS-6.jpeg", as_gray= True)
print(img.shape)
X = img.reshape(-1,1)

# Clustered image
kmeans = KMeans(n_clusters=3)

kmeans.fit(X)

labels = kmeans.labels_
print(min(labels))
print(max(labels))
centroids = kmeans.cluster_centers_
print(centroids)
print(X.shape)

labels = labels.reshape(img.shape)

plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.subplot(122)
plt.imshow(labels, cmap="gray")
plt.show()
```
